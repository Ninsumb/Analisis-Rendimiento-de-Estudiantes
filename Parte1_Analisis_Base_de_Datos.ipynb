#### **Punto 1:** Describe cada columna del conjunto de datos:

* **Horas de Estudio**: Numero de horas que el alumno invirtió en estudiar semanalmente. El dato es un numero entero, y se trata de una variable discreta.

* **Asistencia**: Porcentaje de asistencia del alumno. Es un numero entero en un rango de 60 a 100. Es una variable continua, pero no hay valores intermedios (decimales).

* **Participación de los Padres/Familia**: Nivel de participación de los padres en el desarrollo educativo del estudiante. Es una variable ordinal, que es como una variable categórica pero con un orden, por lo que existe una clasificación para medir (alta, media, baja).

* **Acceso a Recursos**: Disponibilidad de recursos educacionales. También es una variable ordinal (alta, media, baja).

* **Actividades Extracurriculares**: Si el estudiante realiza o no actividades aparte. Variable binaria (si, no).

* **Horas de Sueño**: Promedio de horas de sueño por noche. Varible discreta.

* **Puntajes previos**: Puntuación de examenes anteriores, se interpreta más bien como el promedio del estudiante. Es una variable continua que va de 50 a 100.

* **Nivel de Motivación**: El nivel de motivación del estudiante. Variable ordinal (alta, media, baja).

* **Acceso a Internet**: Es la disponibilidad del acceso a internet del alumno. Variabla binaria (si, no).

* **Clases Particulares**: Numero de clases particulares al que se atendió por mes. Variable discreta.

* **Ingresos Familiares**: El ingreso ecónomico de la familia del estudiante (salarios, sueldos, etc). Variable ordinal (alta, media, baja).

* **Calidad del Profesor**: Calidad de la enseñanza y los recursos didácticos del profesor. Otra variable ordinal (alta, media, baja).

* **Tipo de Escuela**: Tipo de escuela a la que atiende el alumno (pública, privada), será una variable categórica (sin orden ni magnitud).

* **Influencia de los Compañeros**: Influencia del resto de esutdiantes sobre el rendimiento academico. Variable ordinal (positiva, neutral, negativa).

* **Actividad Física**: Promedio entero de las horas en las que el estudiante hace algún tipo de actividad física en la semana. Variable discreta.

* **Discapacidades en el Aprendizaje**: Presencia de discapacidades que puedan tener un impacto en el aprendizaje. Variable binaria (si, no).

* **Nivel Educacional de los Padres**: El nivel educativo más alto alcanzado por los padres del estudiante. Variable categórica (secundaria, universidad, posgraduación).

* **Distancia desde el Hogar**: Distancia desde el hogar a la escuela. Variable ordinal (cerca, moderada, lejos).

* **Género**: Variable categórica (Masculino, femenino).

* **Nota del Examen**: Nota final del examen. Variable continua. //Duda respecto a la base de datos en sí --> El valor max es 101?

#### **Punto 5**: Transformaciones Preliminares:

Ahora voy a normalizar todos las columnas de caracterísitcas, con el objetivo de mejorar el rendimiento del modelo. Para esto voy a modificar el df creado en el punto anterior, pero primero voy a crear dos csv distintos según el método de normalización, y luego evaluaré cada uno por separado junto con los distintos métodos para tratar los datos atípicos. Con el que obtenga la mejor puntuación de accuracy, actualizare el csv que finalmente usaré en mi red. 

Arrancamos con MinMax:

```python
def metodo_MinMax(df):
    
    df_norm = df.iloc[:, :-1].copy()  # Copio todas las columnas de característicasen en el df donde voy a devolver todos los datos normalizados
    mins = df_norm.min() #Con estas funciones, saco el minimo y maximo de cada respectiva columna 
    maxs = df_norm.max()

    
    df_norm = (df_norm - mins) / (maxs - mins)# La ecuación para normalizar 
    
    
    df_norm[df.columns[-1]] = df.iloc[:, -1].values #df.columns[-1] es mi columna objetivo, que antes deje de lado. Ahora, la estoy agregando a mi df normalizado.
    #En la columna "Exam_Score" de mi df normalizado, irán los valores de la misma columna en mi df original 

    return df_norm

df_minmax = metodo_MinMax(df_nuevo)
df_minmax
```

Solo para asegurarme, quiero ver que valores obtendría con la función que ya trae Scikit-learn. 

```python
from sklearn.preprocessing import MinMaxScaler

inputs = df_nuevo.iloc[:, :-1] 
outputs = df_nuevo.iloc[:, -1] 

scaler = MinMaxScaler()

in_norm = scaler.fit_transform(inputs)

df_norm = pd.DataFrame(in_norm, columns=inputs.columns)

df_nuevo = pd.concat([df_norm, outputs.reset_index(drop=True)],axis=1)

df_nuevo
```

¡Me dan lo mismo!

Seguimos con el método estándar:

```python
def norm_estandar(df):

    df_norm = df.iloc[:, :-1].copy() 

    media = df_norm.mean()
    std = df_norm.std()
    
    df_norm = (df_norm - media) / std #Hago exactamente lo mismo que en el método anterior, solo cambio la ecuación.
    
    
    df_norm[df.columns[-1]] = df.iloc[:, -1].values
    return df_norm

df_standard = norm_estandar(df_nuevo)
df_standard
```

Y ahora pruebo con Sklearn...

```python
from sklearn.preprocessing import StandardScaler

inputs = df_nuevo.iloc[:, :-1] #Tomo todas las columnas menos la ultima
outputs = df_nuevo.iloc[:, -1] #Tomo la ultima columna

scaler = StandardScaler()

in_norm = scaler.fit_transform(inputs)

df_norm = pd.DataFrame(in_norm, columns=inputs.columns)

df_nuevo = pd.concat([df_norm, outputs.reset_index(drop=True)],axis=1)

df_nuevo
```

Con esto, voy a empezar las preparaciones para ver qué método me conviene más para tratar a los outliers. 

Primero, vamos a deshacernos de la columna que tiene muchos datos atípicos. Tutoring_Sessions tiene un total de 430 outliers, como vimos en el punto 4. Además, su correlación es de 0.12, por lo que puede considerarse no tan importante para el modelo. Hours_Studied, en cambio, presenta muchos menos outliers (alrededor de 40), y sacarla está fuera de discución ya que es la segunda columna con correlación más fuerte. 

```python
df_pruebaMM_1 = df_minmax.drop(columns=['Tutoring_Sessions'])
df_pruebaS_1 = df_standard.drop(columns=['Tutoring_Sessions'])
```

Luego, vamos a eliminar directamente todas las filas del dataframe que contengan outliers. 

```python
outliers = np.concatenate((datos_atipicos_altos,datos_atipicos_bajos,datos_atipicos)) #Estos arrays con posiciones están definidos en el punto 4


df_pruebaMM_2 = df_minmax.drop(outliers)
df_pruebaS_2 = df_standard.drop(outliers)
```

Por último, vamos a transformar todos los datos atípicos en la media de sus respectivas columnas. 

```python
df_pruebaMM_3 = df_minmax

df_stats1 = df_minmax['Hours_Studied'].describe()
df_stats2 = df_minmax['Tutoring_Sessions'].describe()


df_pruebaMM_3.loc[(datos_atipicos_altos), 'Hours_Studied'] = df_stats1['mean']
df_pruebaMM_3.loc[(datos_atipicos_bajos),'Hours_Studied'] = df_stats1['mean']
df_pruebaMM_3.loc[(datos_atipicos),'Tutoring_Sessions'] = df_stats2['mean']
```

```python
df_pruebaS_3 = df_standard

df_stats1 = df_standard['Hours_Studied'].describe()
df_stats2 = df_standard['Tutoring_Sessions'].describe()

df_pruebaS_3.loc[(datos_atipicos_altos), 'Hours_Studied'] = df_stats1['mean']
df_pruebaS_3.loc[(datos_atipicos_bajos),'Hours_Studied'] = df_stats1['mean']
df_pruebaS_3.loc[(datos_atipicos),'Tutoring_Sessions'] = df_stats2['mean']
```

Ahora sí, teniendo todo listo, vemos qué nos da el accuracy con sklearn: 

```python
from sklearn.model_selection import train_test_split

def split_data(df):
    X = df.values[:, :-1]  
    y = df.values[:, -1]
    return train_test_split(X, y, test_size=1/3,random_state=42)

#Primero, los dos metodos de normalizacion aplicados al df sin modificar los outliers 

X_train1, X_test1, y_train1, y_test1 = split_data(df_minmax)
X_train2, X_test2, y_train2, y_test2 = split_data(df_standard)

#Luego, sin la columna Tutoring_Sessions

X_train3, X_test3, y_train3, y_test3 = split_data(df_pruebaMM_1)
X_train4, X_test4, y_train4, y_test4 = split_data(df_pruebaS_1)

#Sacando las filas con outliers 

X_train5, X_test5, y_train5, y_test5 = split_data(df_pruebaMM_2)
X_train6, X_test6, y_train6, y_test6 = split_data(df_pruebaS_2)

#Cambiando los outliers por la media de la columna 

X_train7, X_test7, y_train7, y_test7 = split_data(df_pruebaMM_3)
X_train8, X_test8, y_train8, y_test8 = split_data(df_pruebaS_3)
```

```python
from sklearn.neural_network import MLPClassifier

def evaluate_model(X_train, y_train, X_test, y_test):
    model = MLPClassifier(solver='sgd',
                   hidden_layer_sizes=(2, ),
                   activation='relu',
                   max_iter=100_000,
                   learning_rate_init=.05,
                   random_state=42)
    model.fit(X_train, y_train)
    accuracy = model.score(X_test, y_test)
    return accuracy

accuracy1 = evaluate_model(X_train1, y_train1, X_test1, y_test1)
accuracy2 = evaluate_model(X_train2, y_train2, X_test2, y_test2)
accuracy3 = evaluate_model(X_train3, y_train3, X_test3, y_test3)
accuracy4 = evaluate_model(X_train4, y_train4, X_test4, y_test4)
accuracy5 = evaluate_model(X_train5, y_train5, X_test5, y_test5)
accuracy6 = evaluate_model(X_train6, y_train6, X_test6, y_test6)
accuracy7 = evaluate_model(X_train7, y_train7, X_test7, y_test7)
accuracy8 = evaluate_model(X_train8, y_train8, X_test8, y_test8)


print(f'Accuracy de Minmax: {accuracy1}')
print(f'Accuracy Standard: {accuracy2}')
print(f'Accuracy MM Prueba 1: {accuracy3}')
print(f'Accuracy S Prueba 1: {accuracy4}')
print(f'Accuracy MM Prueba 2: {accuracy5}')
print(f'Accuracy S Prueba 2: {accuracy6}')
print(f'Accuracy MM Prueba 3: {accuracy7}')
print(f'Accuracy S Prueba 3: {accuracy8}')
```

Cambiando un par de veces el random_State, llegué a la conclusión de que el Accuracy del df normalizado con método de escalado estándar y las filas de outliers eliminadas es el más alto, manteniendose casi siempre en 92%.
