#### **Punto 1:** Describe cada columna del conjunto de datos: 

* **Horas de Estudio**: Numero de horas que el alumno invirtió en estudiar semanalmente. El dato es un numero entero, y se trata de una variable discreta.

* **Asistencia**: Porcentaje de asistencia del alumno. Es un numero entero en un rango de 60 a 100. Es una variable continua, pero no hay valores intermedios (decimales). 

* **Participación de los Padres/Familia**: Nivel de participación de los padres en el desarrollo educativo del estudiante. Es una variable ordinal, que es como una variable categórica pero con un orden, por lo que existe una clasificación para medir (alta, media, baja).

* **Acceso a Recursos**: Disponibilidad de recursos educacionales. También es una variable ordinal (alta, media, baja).

* **Actividades Extracurriculares**: Si el estudiante realiza o no actividades aparte. Variable binaria (si, no).

* **Horas de Sueño**: Promedio de horas de sueño por noche. Varible discreta. 

* **Puntajes previos**: Puntuación de examenes anteriores, se interpreta más bien como el promedio del estudiante. Es una variable continua que va de 50 a 100. 

* **Nivel de Motivación**: El nivel de motivación del estudiante. Variable ordinal (alta, media, baja).

* **Acceso a Internet**: Es la disponibilidad del acceso a internet del alumno. Variabla binaria (si, no).

* **Clases Particulares**: Numero de clases particulares al que se atendió por mes. Variable discreta. 

* **Ingresos Familiares**: El ingreso ecónomico de la familia del estudiante (salarios, sueldos, etc). Variable ordinal (alta, media, baja).

* **Calidad del Profesor**: Calidad de la enseñanza y los recursos didácticos del profesor. Otra variable ordinal (alta, media, baja).

* **Tipo de Escuela**: Tipo de escuela a la que atiende el alumno (pública, privada), será una variable categórica (sin orden ni magnitud).

* **Influencia de los Compañeros**: Influencia del resto de esutdiantes sobre el rendimiento academico. Variable ordinal (positiva, neutral, negativa).

* **Actividad Física**: Promedio entero de las horas en las que el estudiante hace algún tipo de actividad física en la semana. Variable discreta.

* **Discapacidades en el Aprendizaje**: Presencia de discapacidades que puedan tener un impacto en el aprendizaje. Variable binaria (si, no). 

* **Nivel Educacional de los Padres**: El nivel educativo más alto alcanzado por los padres del estudiante. Variable categórica (secundaria, universidad, posgraduación). 

* **Distancia desde el Hogar**: Distancia desde el hogar a la escuela. Variable ordinal (cerca, moderada, lejos).

* **Género**: Variable categórica (Masculino, femenino).

* **Nota del Examen**: Nota final del examen. Variable continua. //Duda respecto a la base de datos en sí --> El valor max es 101?

#### **Punto 2:** Análisis de Correlaciones: 

Antes de hacer el punto 2, tendré que saltar a completar parte de lo que pide el **Punto 5**(Transformaciones Preliminares). Esto se debe a que para ver y analizar las correlaciones tendré que estar trabajando con datos númericos. Lo primero que haré es simplificar los datos "complejos" (todas las variables que no sean númericas) en diccionarios. 


```python
par_involvment = {'Low': 0, 'Medium': 1, 'High': 2}

acc_resources = {'Low': 0, 'Medium': 1, 'High': 2}

extrc_activities = {'No': 0, 'Yes': 1}

motv_level = {'Low': 0, 'Medium': 1, 'High': 2}

internet_acc = {'No': 0, 'Yes': 1}

fam_income = {'Low': 0, 'Medium': 1, 'High': 2}

teacher_qlty = {'Low': 0, 'Medium': 1, 'High': 2} #

sch_type = {'Public': 0, 'Private': 1}

peer_infl = {'Negative': 0, 'Neutral': 1, 'Positive': 2}

learning_dis = {'No': 0, 'Yes': 1}

parental_edu = {'High School': 0, 'College': 1, 'Postgraduate': 2} #

dist = {'Near': 0, 'Moderate': 1, 'Far': 2} #Por algun motivo TODAS las columnas con # devuelven numero flotante

gender = {'Male': 0, 'Female': 1}

```

Teniendo ya listos los diccionarios para la nueva base de datos simplificada, prosigo con el reemplazo. 

```python
#Creamos otro diccionario para agrupar los anteriores, cada uno con la key de su correspondiente columna
remplazo = {
    'Parental_Involvement':par_involvment,
    'Access_to_Resources':acc_resources,
    'Extracurricular_Activities':extrc_activities,
    'Motivation_Level':motv_level,
    'Internet_Access':internet_acc,
    'Family_Income':fam_income,
    'Teacher_Quality':teacher_qlty,
    'School_Type':sch_type,
    'Peer_Influence':peer_infl,
    'Learning_Disabilities':learning_dis,
    'Parental_Education_Level':parental_edu,
    'Distance_from_Home':dist,
    'Gender': gender
}

#Ahora, vamos a iterar sobre cada columna
#"valor_simple" representa el diccionario con el que vamos a reemplazar los valores de la columna
for columna, valor_simple in remplazo.items():
    if columna in df.columns: #Esta linea nos la podriamos ahorrar, pero nos sirve por hubiera algun error en los nombres de las columnas definidas en 'reemplazo'
        df[columna] = df[columna].replace(valor_simple) #Acá ocurre el reemplazo



df
```

Ya con los datos transformados en valores simples, podemos ver la correlación entre las variables: 

```python
import seaborn as sns 
import matplotlib.pyplot as plt

matriz_corr = df.corr() #Calculamos la matriz de correlacion para todas las columanas del df

plt.figure(figsize=(12,12))
sns.heatmap(matriz_corr,annot=True,fmt=".2f",cmap='plasma')#Creamos un mapa de calor de la matriz de correlación
plt.title("Mapa de Correlación")

plt.show()
```

A simple vista no vemos datos anormales teniendo en cuenta nuestras variables iniciales. Naturalmente, el promedio y la desviacion estándar cambiarán cuando llegemos al punto 5 (donde vamos a normalizar algunas columnas). 

Todo parece normal en tanto tomemos en cuenta los distintos rangos de valores que tienen las columnas. A simple vista, ya que no hay conflicto entre la mediana y el promedio de cada columna, parecería que no tendremos problemas con valores atípicos. 

Sacó las columnas que no me interesan tanto (con correlacion respecto a la columna objetivo tendiendo a 0) y voy a dejar las que usaré para la red. 

```python
#Con drop() eliminaremos las columnas que no se van a tomar en cuenta
df_nuevo = df.drop(columns=['Extracurricular_Activities','Sleep_Hours','Motivation_Level','Internet_Access','Family_Income','Teacher_Quality','School_Type','Peer_Influence','Physical_Activity','Learning_Disabilities','Parental_Education_Level','Distance_from_Home','Gender'])
df_nuevo.to_csv("Factores_de_Buen_Rendimiento_en_Estudiantes(actualizado).csv",index=False) #Guardamos el df actualizado en un csv aparte, y sacamos el índice de las filas

df_nuevo.describe()
```

Descarté todas las columnas menores o iguales al módulo de 0,10. Ya preparé la base de datos actualizada, y tengo bastante seguridad acerca de la ausencia de outliers. Lo que voy a hacer a continuación es evaluar las columnas con un gráfico de distribución normal.

```python
import matplotlib.pyplot as plt 
import numpy as np 
from scipy import stats 

for columna in df_nuevo.columns: #Vamos a iterar por cada columna 

    #Primero guardamos en otro df las estadísitcas descriptivas de la columna, y procedemos a guardar en variables las que nos interesan
    df_stats = df[columna].describe()

    media = df_stats['mean']
    mediana = df_stats['50%']
    Q1 = df_stats['25%']
    Q3 = df_stats['75%']
    dev_std = df_stats['std']

    #Creamos los ejes del gráfico
    x = np.linspace(df[columna].min() - 10, df[columna].max() + 10, 1000)#El graf en el eje x se va a manejar entre el minimo de la columna y el maximo, los 10 que se suman y se restan son para ajustar mejor el grafico. El ultimo parametro es la cantidad de puntos   
    y = stats.norm.pdf(x, media, dev_std) #Calcula la funcion 

    plt.figure(figsize=(10,10))

    plt.plot(x, y,color='blue')

    plt.axvline(media, color='orange', linestyle='--', label='Media')
    plt.axvline(mediana, color='green', linestyle='--', label='Mediana')
    plt.axvline(Q1, color='red', linestyle='--', label='Q1 (25%)')
    plt.axvline(Q3, color='purple', linestyle='--', label='Q3 (75%)')

    plt.title(f'Gráfico de Distribución Normal para {columna}')
    plt.xlabel('Valores')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid()

    
    plt.show()
```

No vemos nada raro en los valores, ningún indicio de valores atípicos. En alguna de las columnas, la asíntota verde de la Mediana no aparece porque tiene alguna otra asíntota superpuesta (Q1 o Q3). Como es esperado, esto pasa en las dos únicas columnas que contemplan un rango de 0 a 2 (las que originalmente eran variables ordinales): Parental_Involvement y Access_to_Resources. También ocurre en Tutoring_Sessions, donde el Q1 y la mediana son ambas 1 (y el max son 8 horas), pero tenemos una media cercana a 1. En Attendance la mediana está tan cerca de la media que también se superpone, e igualmente en Hours_Studied la media es apenas visible. 

#### **Punto 5**: Transformaciones Preliminares:

Ahora voy a normalizar todos las columnas de caracterísitcas, con el objetivo de mejorar el rendimiento del modelo. Para esto voy a modificar el df creado en el punto anterior. 

```python
from sklearn.preprocessing import MinMaxScaler

inputs = df_nuevo.iloc[:, :-1] #Tomo todas las columnas menos la ultima
outputs = df_nuevo.iloc[:, -1] #Tomo la ultima columna

scaler = MinMaxScaler()

in_norm = scaler.fit_transform(inputs)

df_norm = pd.DataFrame(in_norm, columns=inputs.columns)

df_nuevo = pd.concat([df_norm, outputs.reset_index(drop=True)],axis=1)

df_nuevo                
```
